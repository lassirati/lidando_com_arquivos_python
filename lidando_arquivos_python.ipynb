```python
import pandas as pd
```

```python
# lendo o csv completo
# (este é pequeno, portanto é simples de ler.
# mas ele poderia ser enorme e nao termos memoria suficiente)
pd.read_csv('Pasta1.csv', sep=';')
```

```python
# queremos essa situação, a soma de todas as moedas
# se o dataframe fosse enorme nao conseguiriamos sequer ler,
# menos ainda fazer o groupby
pd.read_csv('Pasta1.csv', sep=';').groupby('moeda').sum().reset_index()
```

```python
# podemos ler entao em chunkcs de tamanho 5
# aqui leria-se 4 chunks de 5 dados, totalizando os 20 dados da tabela
# mas eu coloquei um BREAK após a primeira leitura, apenas pra compreender o que acontece
# nota-se que houve um groupby desse pequeno chunk, que se somormos as respectivas moedas
# das 5 primeiras linhas da tabelona veriamos que o valor bate pra essa agregacao
for chunk in pd.read_csv('Pasta1.csv', sep=';', chunksize=5):
    resumo = chunk.groupby("moeda").sum()
    display(resumo)
    break
```

```python
# aqui fiz o mesmo pra ficar evidente o que acontecerá, fiz um contador pra ler 2 chunks e depois haver um break
# nota-se que ele gera duas tabelas de agregacao (uma das 5 primeiras linhas, das proximas 5 linhas)
contador = 0
for chunk in pd.read_csv('Pasta1.csv', sep=';', chunksize=5):
    resumo = chunk.groupby("moeda").sum()
    print('resumo das {}as 5 linhas'.format(contador+1))
    display(resumo)
    contador = contador+1
    if contador > 1:
        break
```

```python
# entao eu preciso criar um DF vazio e ir apendando
# o que no fim nos gera um grande dataframe de 4 agregacoes de 5 linhas
tabela = pd.DataFrame()
for chunk in pd.read_csv('Pasta1.csv', sep=';', chunksize=5):
    resumo = chunk.groupby("moeda").sum().reset_index()
    tabela = pd.concat([tabela, resumo])
    
tabela    
```

```python
# se eu agregar a tabela de agregacoes, obtenho o resultado
# que me interessa que é a soma das moedas da base de dados inteira
tabela_final = tabela.groupby("moeda").sum().reset_index()
tabela_final
```

```python
import pandas as pd
from glob import glob
```

```python
## listando todos os arquvivos na pasta raiz com ext .xlsx
## se tivesse numa outra pasta seria arquivos = glob("Nome_da_Pasta/*.xlsx")
arquivos = glob("*.xlsx")
arquivos
```

```python
## criando uma lista_aux para ir lendo os dataframes e concatenando

lista_dfs = []
for arquivo in arquivos:
    lista_dfs.append(pd.read_excel(arquivo))
todos_arquivos = pd.concat(lista_dfs).reset_index(drop=True)
todos_arquivos
```

```python
## o mesmo mas de foram mais direta, sem lista_aux para os DFs

todos_arquivos = pd.concat(pd.read_excel(arquivo) for arquivo in arquivos)
todos_arquivos = todos_arquivos.reset_index(drop=True)
todos_arquivos
```
